```python
hitungluas = input("""Pilih salah satu:
1. Luas Persegi
2. Luas Lingkaran
3. Luas Segitiga
""")
match hitungluas:
    case "1":
       print("Menghitung Luas Persegi")
       sisi = int(input("Masukkan nilai sisi: "))
       hitungluaspersegi = sisi**2
       print(f"Jadi luas persegi dengan sisi", sisi, "cm adalah", hitungluaspersegi, "cm^2")
    case "2":
       print("Menghitung Luas Lingkaran")
       r = int(input("Masukkan nilai jari-jari: "))
       tinggi = int(input("Masukkan nilai tinggi: "))
       phi = 22/7
       hitungluaslingkaran = int(2*phi*r*tinggi)+(2*phi*r**2)
       print(f"Jadi luas lingkaran dengan jari-jari", r, "cm adalah", hitungluaslingkaran, "cm^2")
    case "3":
       print("Menghitung Luas Segitiga")
       alas = int(input("Masukkan nilai alas: "))
       tinggi = int(input("Masukkan nilai tinggi: "))
       hitungluassegitiga = alas*tinggi/2
       print(f"Jadi luas segitiga dengan alas", alas, "cm adalah", hitungluassegitiga, "cm^2")
    case _:
       print("Pilihan Tidak Valid!")
```
